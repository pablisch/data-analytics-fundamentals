```python
# TASK 1
# Context - During a recent customer support review, it was discovered that some customer records had incorrect contact numbers, mistakenly starting with “0800” or “0845”, which are often reserved for national helplines or business contact numbers. To avoid miscommunication, you have been tasked with finding all customer records with contact numbers that start with these prefixes.
# Task - Write a query to list the customer’s ID, first name, last name, and contact number for all customers whose contact numbers start with 0800 or 0845.
query = """
SELECT
    id AS customer_id,
    first_name,
    last_name,
    contact_number
FROM Customer
WHERE contact_number LIKE '0800%'
   OR contact_number LIKE '0845%';
        """
run_sql_query(query, cursor)
```

```python
# TASK 1 - ADDING VALIDATION for disjointed numbers
# Context - During a recent customer support review, it was discovered that some customer records had incorrect contact numbers, mistakenly starting with “0800” or “0845”, which are often reserved for national helplines or business contact numbers. To avoid miscommunication, you have been tasked with finding all customer records with contact numbers that start with these prefixes.
# Task - Write a query to list the customer’s ID, first name, last name, and contact number for all customers whose contact numbers start with 0800 or 0845.
query = """
SELECT
    id AS customer_id,
    first_name,
    last_name,
    contact_number
FROM Customer
WHERE REPLACE(contact_number, ' ', '') LIKE '0800%'
   OR REPLACE(contact_number, ' ', '') LIKE '0845%';
        """
run_sql_query(query, cursor)
```

```python
# TASK 2
# Context - The logistics team wants to review order delivery details for tracking and delivery route optimization. They have asked for a list of all orders showing the order date, customer’s name, and the delivery address where the order was sent.
# Task - Write a query to display the order date, customer name (first and last name combined as customer_name), and the house name/number and postcode of the delivery address for each order. Sort the results by order date in ascending order.
query = """
SELECT
    o.order_date,
    CONCAT(c.first_name, ' ', c.last_name) AS customer_name,
    a.house_name_number,
    a.postcode
FROM OnlineOrder o
JOIN Customer c
    ON o.customer_id = c.id
JOIN Address a
    ON o.delivery_address_id = a.id
ORDER BY o.order_date ASC;
        """
run_sql_query(query, cursor)
```

```python
# TASK 3 - returns ALL products
# Context - The marketing team is planning a promotional campaign and wants to focus on the most popular products. You have been asked to generate a list of products, along with the total number of times each product was ordered. The list should be sorted from the most to the least popular product.
# Task - Write a query to display the product title and the total number of times it was ordered. Sort the results by the total number of times each product was ordered in descending order.
query = """
SELECT
    p.title,
    COALESCE(SUM(oc.amount), 0) as amount
FROM Product p
LEFT JOIN OrderContent oc
    ON oc.product_id = p.id
GROUP BY p.title
ORDER BY COALESCE(SUM(oc.amount), 0) DESC;
        """
run_sql_query(query, cursor)
```

```python
# TASK 3 - only returns ordered products
# Context - The marketing team is planning a promotional campaign and wants to focus on the most popular products. You have been asked to generate a list of products, along with the total number of times each product was ordered. The list should be sorted from the most to the least popular product.
# Task - Write a query to display the product title and the total number of times it was ordered. Sort the results by the total number of times each product was ordered in descending order.
query = """
        SELECT
            p.title,
            SUM(oc.amount) AS amount
        FROM Product p
        JOIN OrderContent oc
            ON p.id = oc.product_id
        GROUP BY
            p.title
        ORDER BY
            SUM(oc.amount) DESC;
        """
run_sql_query(query, cursor)
```

```python
# TASK 4
# Context - The customer service team needs to identify customers who have more than one registered address in the system. This is important for flagging potential duplicates or handling complex delivery requirements.
# Task - Write a query to find all customers who have more than 1 registered address. Display the first name, last name, and number of addresses they have registered. Only list customers with more than one address.
query = """
SELECT
    c.first_name,
    c.last_name,
    COUNT(a.id) AS number_of_addresses
FROM Customer c
JOIN Address a
    ON a.customer_id = c.id
GROUP BY
    c.id,
    c.first_name,
    c.last_name
HAVING COUNT(a.id) > 1;
        """
run_sql_query(query, cursor)
```

```python
# TASK 5
# Context - The finance team is preparing a customer sales report to identify high-value customers for a loyalty program. You have been asked to provide the total revenue generated by each customer from all of their orders.
# Task - Write a query to calculate the total revenue for each customer and display the top 3. Display the customer’s full name (as customer_name) and their total revenue from all orders. The results should be sorted by total revenue in descending order, showing the most valuable customers first.
query = """
SELECT
    CONCAT(c.first_name, ' ', c.last_name) AS customer_name,
    SUM(oc.amount * p.price) AS total_revenue
FROM Customer c
     JOIN OnlineOrder o
          ON c.id = o.customer_id
     JOIN OrderContent oc
          ON o.id = oc.order_id
     JOIN Product p
          ON oc.product_id = p.id
GROUP BY c.id, c.first_name, c.last_name
ORDER BY total_revenue DESC
    LIMIT 3;
        """
run_sql_query(query, cursor)
```
