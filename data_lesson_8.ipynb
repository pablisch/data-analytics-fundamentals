```python
def run_query(query):
    cursor.execute(query)
    rows = cursor.fetchall()
    columns = [desc[0] for desc in cursor.description]
    df = pd.DataFrame(rows, columns=columns)
    return df
```

```python
# Aggregation - COUNT
query = """SELECT customer_id, COUNT(*) AS total_orders FROM OnlineOrder GROUP BY customer_id HAVING total_orders > 2;"""
run_query(query)
```

```python
# Aggregation - AVERAGE
query = """SELECT category, AVG(price) AS avg_price FROM Product GROUP BY category HAVING avg_price > 3;"""
run_query(query)
```
