```python
import mysql.connector
import pandas as pd
from dotenv import load_dotenv
import os
from sql_helpers import run_sql_query, create_lesson_8_db
load_dotenv()  # loads variables from .env file
mysql_password = os.getenv("MYSQL_PASSWORD")
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password=mysql_password,
)
cursor = connection.cursor()
```

```python
create_lesson_8_db(cursor, connection)
```

```python
# Aggregation - COUNT
query = """SELECT customer_id, COUNT(*) AS total_orders FROM OnlineOrder GROUP BY customer_id HAVING total_orders > 2;"""
run_sql_query(query, cursor)
```

```python
# Aggregation - AVERAGE
query = """SELECT category, AVG(price) AS avg_price FROM Product GROUP BY category HAVING avg_price > 3;"""
run_sql_query(query, cursor)
```
